```python
df = df[["DATE", "ZIP CODE", "# OF NEW", "# OF GOOD", " # OF UA", "# OF PUR", "SPORT"]]
```

```python
df
```

```python
def string_to_float(number):
    if type(number) is float or type(number) is int:
        return number
    if "," in number:
        number = number.replace(",", "")
        return float(number)
    else:
        return float(number)    
```

```python
df["# OF NEW"] = df["# OF NEW"].map(string_to_float)
df["# OF GOOD"] = df["# OF GOOD"].map(string_to_float)
df[" # OF UA"] = df[" # OF UA"].map(string_to_float)
df["# OF PUR"] = df["# OF PUR"].map(string_to_float)
```

```python
df["Given"] = df["# OF NEW"] + df["# OF GOOD"] + df[" # OF UA"] + df["# OF PUR"]
```

```python
sport_df = df[["SPORT", "Given"]].dropna()
```

```python
translate_dict = {"RUN": "RUNNING", "BASETBALL": "BASKETBALL", "BASBALL": "BASEBALL", "UA BASEBALL": "BASEBALL", "UA BB": "BASEBALL", "FB": "FOOTBALL", "UA FOOTBALL": "FOOTBALL", "UA FB CLEATS": "FOOTBALL", "MISC." : "MISC", "NA": "MISC", "MISCELLANEOUS": "MISC", "VB": "VOLLEYBALL", "SB": "SOFTBALL", "UA SB": "SOFTBALL", "lacross": "LACROSSE"}
frequency = {"RUNNING":0, "SOCCER":0, "BASKETBALL":0,"BASEBALL":0,"FOOTBALL":0, "MISC":0,"TRACK":0,"VOLLEYBALL":0,"DANCE":0,"GOLF":0,"SOFTBALL":0,"WRESTLING":0,"LACROSSE":0,"BIKE":0,"TENNIS":0,"CHEER":0,"THUNDER":0,"CLOTHING":0,"GYM":0,"SWIM":0}
total_count = 0
for row in range(len(sport_df)):
    series = sport_df.iloc[row]
    sport = str(series["SPORT"]).strip().upper()
    if sport in translate_dict:
        sport = translate_dict[sport]
    if sport in frequency:
        frequency[sport] += float(series["Given"])
        total_count += float(series["Given"])
```

```python
for key in frequency.keys():
    frequency[key] = frequency[key]/total_count
```
